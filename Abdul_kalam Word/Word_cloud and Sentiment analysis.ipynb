```python
import re
import string 
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
```

```python
with open('C:/Nanduzz/sTudY/ML/Machine Learning/DeepLearning/NLP/project/Abdul_Kalam.txt','r',encoding='utf-8') as f:
    txt=f.readlines()
```

```python
txt
```

```python
txt=[i.strip()for i in txt]
print(txt)
```

```python
txt1=' '.join(txt)
print(txt1)
```

```python
#LOWER
txt2=txt1.lower()
print(txt2)
```

```python
#cleaning the punctuations 
txt3=re.sub(r'[^\w\s]','',txt2)
print(txt3)
```

```python
#cleaning the digits
txt4=re.sub(r'\d+',' ',txt3)
print(txt4)
```

```python
#stopwords
token=word_tokenize(txt4)
token
```

```python
txt5=[i for i in token if i not in stopwords.words('english')]
print(txt5)
```

```python
txt5=' '.join(str(x) for x in txt5)
print(txt5)
```

```python
#remove single letters
txt6=' '.join(i for i in txt5.split() if not (i.isalpha() and len(i)==1) )
txt6
```

```python
#white spaces
txt6=txt6.strip()
txt6
```

```python
words_dict={}
for i in txt6.split():
    words_dict[i]=words_dict.get(i,0)+1
for j in sorted(words_dict):
    print('{}:{}'.format(j,words_dict[j]))
```

```python
word_cloud=WordCloud(width=480,height=480,margin=0).generate(txt6)
```

```python
import matplotlib.pyplot as plt
plt.imshow(word_cloud,interpolation='bilinear')
plt.axis('off')
plt.margins(x=0,y=0)
plt.show()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
analyse=SentimentIntensityAnalyzer()
```

```python
score=analyse.polarity_scores(txt6)
score
```

```python
for i in sorted(score):
    print('{0}:{1} '.format(i,score[i]),end='')
```
